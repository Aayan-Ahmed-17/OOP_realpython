### Defining a Class and Initializing Instance Attributes
This section demonstrates how to define a Python class and initialize its instance attributes using the `__init__` method.

```python
"""
1) Use Captalized Case for classs naming convention

init():

sets the initial state of the object by assigning the values of the object’s properties.
Attributes created in it are called instance attributes
"""
class Dog():
  def __init__(self, name, age):
    self.name = name
    self.age = age
```

### Class Attributes, Instantiation, and Accessing Instance Attributes
This section illustrates the use of class attributes, how to create instances (objects) of a class, and how to access their instance attributes.

```python
class Dog():
  species = "Canis familiaris"

  def __init__(self, name, age):
    self.name = name
    self.age = age

#instantiating a class
miles = Dog("Miles", 4)
buddy = Dog("Buddy", 9)

#accessing instances attrs
print(f"{miles.name} is {miles.age} yrs old", f"{buddy.name} is {buddy.age} yrs old", sep='\n')
```

### Modifying Instance and Class Attributes
This section shows how to change the values of instance attributes and how assigning to an instance's class attribute actually creates a new instance attribute.

```python
# changing vals
buddy.age = 10
print(buddy.age)


miles.species = "Felis silvestris"
print(miles.species)
```

### Instance Methods and Calling Methods
This section demonstrates how to define instance methods within a class and how to call these methods on class instances.

```python
#Other instance method

class Dog():
  species = "Canis familiaris"

  def __init__(self, name, age):
    self.name = name
    self.age = age

  # instance method
  def __str__(self):
    return f"{self.name} is {self.age} yrs old"

  # another instance method
  def speak(self, sound):
    return f"{self.name} says {sound}"

"""------------------------------------------------------------"""
#instantiating
miles = Dog("Miles", 4)

#accessing instance methods and attr
print(miles)
print(miles.speak("Woof Woof"))
print(miles.speak("Bow Wow"))
```

### **Exercise**
Create a Car class with two instance attributes:

.color, which stores the name of the car’s color as a string
.mileage, which stores the number of miles on the car as an integer
Then create two Car objects—a blue car with twenty thousand miles and a red car with thirty thousand miles—and print out their colors and mileage. Your output should look like this:

=> The blue car has 20,000 miles
=> The red car has 30,000 miles

```python
class Car():
  def __init__(self, color, mileage):
    self.color = color
    self.mileage = mileage

  def __str__(self):
    return f"The {self.color} car has {self.color} miles"

blue_car = Car(color="blue", mileage=20_000)
red_car = Car(color="red", mileage=30_000)

#printing through tuple
for car in (blue_car, red_car):
  print(car)
```

### Basic Inheritance Example
This section provides a basic example of inheritance, where a `Child` class inherits attributes and methods from a `Parent` class. This demonstrates the fundamental concept of code reuse through class hierarchies.

```python
#basic eg of inheritance
class Parent:
    hair_color = "brown"

class Child(Parent):
  pass
```

**Dog's Child classes**
To add breed for each dog. we'll create child classes

```python
class JackRussellTerrier(Dog):
    pass

class Dachshund(Dog):
    pass

class Bulldog(Dog):
    pass

# instantiating Dog class using child classes
miles = JackRussellTerrier("Miles", 4)
buddy = Dachshund("Buddy", 9)
jack = Bulldog("Jack", 3)
jim = Bulldog("Jim", 5)

#accessing instances attributes and methods
print(miles.species)

print(buddy.name)

print(jack)

print(jim.speak("Woof"))
```

Find relation between instance and class using:


1.   type(instance)
2.   ininstance(instance, Class)



```python
print(type(miles))
print(isinstance(miles, Dog))
print(isinstance(miles, Bulldog))
print(isinstance(jack, Dachshund))
```

**Override a Parent class method**:
every breed have slightly different sound, you want to provide a default value for the sound argument of their respective .speak() methods.

```python
class JackRussellTerrier(Dog):
    def speak(self, sound="Arf"):
        return f"{self.name} says {sound}"

miles = JackRussellTerrier("Miles", 4)
print(miles.speak())
print(miles.speak("Grrr"))
```

Making changes in Parent class reflect in child class

```python
class Dog():
  species = "Canis familiaris"

  def __init__(self, name, age):
    self.name = name
    self.age = age

  # instance method
  def __str__(self):
    return f"{self.name} is {self.age} yrs old"

  # another instance method
  def speak(self, sound):
    return f"{self.name} barks: {sound}"

jim = Bulldog("Jim", 5)
print(jim.speak("Woof"))

# but changes would not reflect in JackRussellTerrier class bcz it explicity modified this method
miles = JackRussellTerrier("Miles", 4)
print(miles.speak())
```
